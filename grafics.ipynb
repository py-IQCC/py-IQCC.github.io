## 3.1 Conceptes bàsics de la representació gràfica

Abans de crear la nostra primera representació gràfica, necessitem algunes dades per representar-la, de manera que generarem punts de dades a partir de funcions d'ona radial orbital. L'equació següent defineix la funció d'ona ($\psi$) per a l'orbital atòmic 3s de l'hidrogen respecte al radi atòmic ($r$) en Bohrs ($a_0$).

$$ \psi_{3s} = \frac{2}{27}\sqrt{3}(2r^{2/9} - 2r + 3)e^{-r/3} $$

Generarem punts en aquesta corba utilitzant un mètode anomenat comprensió de llistes. En els exemples següents, `r` és la distància des del nucli i `psi_3s` és la funció d'ona. Si trieu representar una altra cosa, només cal que feu dues llistes o tuples de la mateixa longitud que continguin els valors $x$ i $y$.

```python
# create Python function for generating 3s radial wave function
import math

def orbital_3S(r):
    wf = (2/27)*math.sqrt(3)*(2*r**2/9 - 2*r + 3)* math.exp(-
         r/3)
    return wf
```

```python
# generate data to plot
r = [num / 4 for num in range(0, 150, 3)]
psi_3s = [orbital_3S(num) for num in r]
```

### 3.1.1 Primer gràfic

Per visualitzar les funcions d'ona 3s, anomenarem la funció `plot()`, que és una funció d'ús general per a la representació gràfica. Les dades `r` i `psi_3s` s'hi introdueixen com a arguments posicionals com a variables $x$ i $y$, respectivament.

```python
plt.plot(r, psi_3s, 'o');
```

Per defecte, matplotlib crea un diagrama de dispersió utilitzant el blau com a color predeterminat. Això es pot modificar si els cercles blaus no són del vostre gust. Si la funció `plot()` només té un argument, matplotlib assumeix que les dades són els valors $y$ i els representa gràficament contra els seus índexs.

Si es proporciona una cadena buida a `linestyle` o `marker`, no s'inclou cap línia ni marcador, respectivament, al gràfic. ### 3.1.3 Etiquetes

Sovint és important etiquetar els eixos del gràfic. Això s'aconsegueix mitjançant les funcions `plt.xlabel()` i `plt.ylabel()`, que es col·loquen en línies diferents com la funció `plt.plot()`. Ambdues funcions accepten cadenes de caràcters.

```python
plt.plot(r, psi_3s, 'go-')
plt.xlabel('X Values')
plt.ylabel('Y Values');
```

En cas que vulgueu un títol a la part superior dels vostres gràfics, podeu afegir-ne un utilitzant l'argument `plt.title()`. Per afegir símbols als eixos, això es pot fer mitjançant ordres de LaTeX que s'utilitzen a continuació, però la discussió sobre LaTeX està fora de l'abast d'aquest capítol.

```python
plt.plot(r, psi_3s, 'go-')
plt.xlabel('Radius, Bohrs')
plt.ylabel('Wave Function, $\\Psi$')
plt.title('3S Radial Wave Function');
```

### 3.1.4 Mida de la figura

Si voleu canviar la mida o les dimensions de la figura al quadern Jupyter, ho podeu fer amb `plt.figure(figsize=(width, height))`. És important que aquesta funció estigui *a sobre* de la funció de gràfic real i no a sota perquè pugui modificar la figura.

```python
plt.figure(figsize=(8,4))
plt.plot(r, psi_3s, 'go-')
plt.xlabel('Radius, Bohrs')
plt.ylabel('Wave Function, $\\Psi$')
plt.title('3S Radial Wave Function');
```

### 3.1.5 Desar figures

La majoria dels usos de matplotlib són per generar figures en un quadern Jupyter. Tanmateix, hi ha vegades que cal desar les figures en fitxers per a un manuscrit, informe o presentació. En aquestes situacions, podeu desar el vostre gràfic mitjançant la funció `plt.savefig()`, que accepta uns quants arguments. El primer i únic argument necessari és el nom del fitxer de sortida com a cadena. A continuació, l'usuari també pot triar la resolució en punts per polzada mitjançant l'argument de paraula clau `dpi`. Finalment, hi ha diversos formats de fitxer compatibles amb les funcions `plt.savefig()`, com ara PNG, TIF, JPG, PDF, SVG, entre d'altres. Els formats es poden seleccionar mitjançant l'argument `format`, que també accepta una cadena, i si no es tria cap format explícitament, matplotlib utilitza PNG per defecte.

```python
plt.plot(r, psi_3s, 'g.-')
plt.savefig('my_image.png', format='PNG', dpi=600);
```

## 3.3 Superposició de gràfics

Sovint cal representar més d'un conjunt de dades als mateixos eixos, i això es pot aconseguir de dues maneres amb matplotlib. La primera és cridar la funció de representació dues vegades a la mateixa cel·la de codi Jupyter. Matplotlib col·locarà automàticament els dos gràfics a la mateixa figura i l'escalarà adequadament per incloure totes les dades. A continuació, les dades de la funció d'ona per a l'orbital d'hidrogen 3p es generen de manera similar a la dels 3s anteriors, de manera que ara les funcions d'ona per als orbitals 3s i 3p es poden representar al mateix conjunt d'eixos.

```python
def orbital_3P(r):
    wf = (math.sqrt(6)*r*(4-(2/3)*r)*math.e**(-r/3))/81
    return wf

r = [num / 4 for num in range(0, 150)]
psi_3p = [orbital_3P(num) for num in r]
psi_3s = [orbital_3S(num) for num in r]
```

```python
plt.plot(r, psi_3s)
plt.plot(r, psi_3p)
plt.xlabel('Radius, Bohrs')
plt.ylabel('Wave Function');
```

El segon mètode és incloure els dos conjunts de dades a la mateixa ordre de gràfics, tal com es mostra a continuació. Matplotlib assumirà que cada nova paraula no clau és un nou conjunt de dades i que els arguments posicionals estan associats a les dades més recents.

```python
plt.plot(r, psi_3s, 'bo', r, psi_3p,'r^')
plt.xlabel('Radius, Bohrs')
plt.xlabel('Radius, Bohrs')
plt.ylabel('Wave Function');
```

En el segon gràfic anterior, `r`, `psi_3s`, `'bo'` són les dades i l'estil del primer conjunt de dades, mentre que `r`, `psi_3p`, `'r^'` són les dades i l'estil de gràfic del segon.

Un problema que sorgeix ràpidament amb els gràfics de diverses figures és identificar quins símbols pertanyen a quines dades. Matplotlib permet a l'usuari afegir una llegenda al gràfic. L'usuari primer ha de proporcionar una etiqueta per a cada conjunt de dades utilitzant l'argument de paraula clau `label=`. Finalment, cridar `plt.legend()` fa que les etiquetes es mostrin al gràfic. Per defecte, matplotlib col·loca la llegenda on decideix que és la ubicació òptima, però aquest comportament es pot anul·lar afegint un argument de paraula clau `loc=`. Una llista completa dels arguments de ubicació està disponible al [lloc web de matplotlib](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html).

També seria útil incloure una línia horitzontal al zero com a guia per a la vista. Matplotlib inclou una funció `plt.hlines(y, xmin, xmax)` precisament per a aquest propòsit, i aquesta funció accepta arguments similars per al color i l'estil.

```python
plt.plot(r, psi_3s, label='3s orbital')
plt.plot(r, psi_3p, label='3p orbital')
plt.hlines(0, 0, 35, linestyle='dashed', color='C3')
plt.xlabel('Radius, Bohrs')
plt.ylabel('Wave Function')
plt.legend();
```
